## Transform

Preparing the dataset for the project purpose

```python
ut.data_review(df_games)
```

```python
df_games_cp1 = df_games.copy()
```

```python
# Removing records with nulls in all of the first 15 columns

df_games_cp1.dropna(subset=df_games_cp1.columns[0:15], how='all', inplace=True)
```

```python
ut.data_review(df_games_cp1)
```

```python
# Removing meaningless null values from the df

df_games_cp1 = df_games_cp1.dropna(subset=['app_name'])
df_games_cp1 = df_games_cp1.dropna(subset=['reviews_url'])
df_games_cp1 = df_games_cp1.dropna(subset=['id'])
```

```python
ut.data_review(df_games_cp1)
```

```python
df_games_cp1[df_games_cp1['title'] != df_games_cp1['app_name']]
```

```python
# Removing meaningless columns from the df (title == app_name // 
# user_id, steam_id, items, items_count, discount_price, metascore with ~100% null values)

df_games_cp1.drop(columns=['title', 'user_id', 'steam_id', 'items', 'items_count', 'discount_price', 'metascore'], inplace=True)
```

```python
ut.data_review(df_games_cp1)
```

```python
df_games_cp1['price'][df_games_cp1['price'].notnull()].unique()
```

```python
words = ['Free', 'Now', 'Demo', 'Install', 'party']

seen = []

for value in df_games_cp1['price'].values:

    try:
        if any(word in value for word in words) and value not in seen:
            seen.append(value)
    except:
        continue

seen
```

```python
df_games_cp1['price'] = df_games_cp1['price'].replace(seen, '0')
```

```python
df_games_cp1['price'] = df_games_cp1['price'].replace('Starting at $499.00', '499.0')
df_games_cp1['price'] = df_games_cp1['price'].replace('Starting at $449.00', '449.0')
```

```python
df_games_cp1['price'][df_games_cp1['price'].notnull()].unique()
```

```python
df_games_cp1['price'] = df_games_cp1['price'].astype(float)
df_games_cp1['id'] = df_games_cp1['id'].astype(int)
```
